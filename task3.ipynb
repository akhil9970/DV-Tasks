```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
file_path = "/content/Stress_Dataset.csv"
df = pd.read_csv(file_path)

# Clean column names
df.columns = df.columns.str.strip()

# Convert Gender if coded numerically
if df["Gender"].dtype in [int, float]:
    df["Gender"] = df["Gender"].map({0: "Male", 1: "Female"})

# Step 3: Identify variables
categorical_vars = ["Gender", "Which type of stress do you primarily experience?"]
continuous_vars = [
    "Have you recently experienced stress in your life?",
    "Have you noticed a rapid heartbeat or palpitations?",
    "Do you face any sleep problems or difficulties falling asleep?",
    "Do you get irritated easily?"
]

# --- Step 4: Bivariate Analysis Plots ---

# 1. Boxplot  Stress vs Gender
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Gender", y=continuous_vars[0])
plt.title("Stress Levels by Gender")
plt.show()

# 2. Violin Plot  Sleep Problems vs Gender
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x="Gender", y=continuous_vars[2])
plt.title("Sleep Problems by Gender")
plt.show()

# 3. Strip Plot  Irritability vs Gender
plt.figure(figsize=(8,5))
sns.stripplot(data=df, x="Gender", y=continuous_vars[3], jitter=True, alpha=0.6)
plt.title("Irritability by Gender (Strip Plot)")
plt.show()

# 4. Swarm Plot  Heartbeat vs Gender
plt.figure(figsize=(8,5))
sns.swarmplot(data=df, x="Gender", y=continuous_vars[1])
plt.title("Heartbeat Issues by Gender (Swarm Plot)")
plt.show()

# 5. Barplot  Avg Stress Level by Stress Type
plt.figure(figsize=(10,5))
sns.barplot(data=df, x="Which type of stress do you primarily experience?", y=continuous_vars[0])
plt.title("Average Stress Levels by Stress Type")
plt.xticks(rotation=45, ha="right")
plt.show()

# 6. Scatter Plot  Stress vs Heartbeat, colored by Gender
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x=continuous_vars[0], y=continuous_vars[1], hue="Gender", alpha=0.7)
plt.title("Stress vs Heartbeat by Gender")
plt.show()

# 7. Correlation Heatmap (continuous variables only)
plt.figure(figsize=(8,6))
sns.heatmap(df[continuous_vars].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Stress Indicators")
plt.show()
```
