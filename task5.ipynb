```python

# Install dependencies (if not already available in Colab)
!pip install plotly pandas

# Import libraries
import pandas as pd
import plotly.express as px

# Load dataset (upload CSV to Colab first or mount Google Drive)
from google.colab import files
uploaded = files.upload()

# Read the uploaded CSV (adjust filename if needed)
df = pd.read_csv("Sport car price.csv")

# Clean the Price column (remove commas and convert to int)
df["Price (in USD)"] = df["Price (in USD)"].str.replace(",", "").astype(int)

# Create tree representation (Treemap)
fig = px.treemap(
    df,
    path=["Car Make", "Car Model"],   # Hierarchy levels
    values="Price (in USD)",          # Block size by price
    color="Horsepower",               # Block color by horsepower
    hover_data={
        "Year": True,
        "Engine Size (L)": True,
        "Torque (lb-ft)": True,
        "0-60 MPH Time (seconds)": True,
        "Price (in USD)": ":,"
    },
    title="Sports Car Dataset - Tree Representation"
)

# Show the plot
fig.show()
```
